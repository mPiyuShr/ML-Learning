## Data visualization

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(10)
y1 = x**2
y2 = (2*x)+3
```

```python
plt.plot(x,y1)           # Pass first x-axix and then value of y-axis
plt.show()               # will print or show your graph
```

```python
# Will print both graphs in one image
plt.plot(x,y1 , color = "Red" , label = "marks")  # will make the line as red
plt.plot(x,y2 , color = "blue" , label = "Fee")
plt.xlabel("X - axis")          # Name of x-axis
plt.ylabel("Y - axis")          # Name of x-axis
plt.title("My Plot")            # name of graph
plt.legend()

plt.show()                  
```

#### Scatter Plots

```python
plt.figure(figsize = (5,3))
plt.scatter(x,y1 , color = "Red" , marker = "*" , label = "Marks")  # will make the line as red
plt.scatter(x,y2 , color = "blue" , label = "Fee" , marker = "o" , linestyle="dashed")
plt.xlabel("X - axis")          # Name of x-axis
plt.ylabel("Y - axis")          # Name of x-axis
plt.title("My Plot")            # name of graph
plt.legend()

plt.show()   
```

### Bar Plot

```python
plt.bar([1,4,3] , [10,20,30])
plt.show()
```

```python
plt.style.use("seaborn")
x = np.array([0,2,4])
plt.bar(x , [10,20,15] , width = 0.5 , label = "This year" , tick_label= ["A" , "B" , "C"])
plt.bar(x+0.5 , [20,10,12] , width = 0.5 , label = "Next year" )
plt.title("Our Plot")
plt.legend()
plt.show()
```

```python
plt.style.use("dark_background")
x = np.array([0,2,4])
plt.bar(x , [10,20,15] , width = 0.5 , label = "This year" , tick_label= ["A" , "B" , "C"])
plt.bar(x+0.5 , [20,10,12] , width = 0.5 , label = "Next year" )
plt.title("Our Plot")
plt.legend()
plt.show()
```

### Pie chart

```python
subs = 'A' , 'B' , 'C' , 'D'
weight = [15,20,35,40]

plt.pie(weight , labels=subs)
plt.show()
```

```python
subs = 'A' , 'B' , 'C' , 'D'
weight = [15,20,35,40]

plt.pie(weight , labels=subs ,explode=(0,0,0.1,0) , shadow=True)
plt.title("Pie-Chart")
plt.show()
```
