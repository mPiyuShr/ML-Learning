### Working with movie data set

```python
md = pd.read_csv("movie_metadata.csv")
```

```python
md
```

```python
titles = md.get('movie_title')
```

```python
print(titles)
```

```python
freq_title = {}

for i in titles:
    length = len(i)
    
    try:
        freq_title[length]+=1
    except:
        freq_title[length]=1

print(freq_title)
```

```python
x = np.array(list(freq_title.keys()))
y = np.array(list(freq_title.values()))
```

```python
plt.scatter(x,y)
plt.xlabel("Title_Length")
plt.ylabel("No. of movies")
plt.title("Movie-length Graph")
plt.show
```
